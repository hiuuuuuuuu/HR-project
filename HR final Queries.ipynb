### 1. What is the gender breakdown of employees in the company?

```python
%%sql
SELECT 
    gender,
    COUNT(*) AS count
FROM 
    hr
WHERE
    termdate ='0000-00-00'
GROUP BY 
    gender;
```

### 2. What is the race/ethnicity breakdown of employees in the company?

```python
%%sql
SELECT 
	race,
    COUNT(*) AS count_race
FROM
	hr
WHERE
	termdate = '0000-00-00'
GROUP BY
	race
ORDER BY 
	count_race DESC;
```

### 5. What is the average length of employment for employees who have been terminated?

```python
%%sql

SELECT
	ROUND(AVG(datediff(termdate,hire_date))/365,1) as avg_length_employee
FROM 
	hr
WHERE
	termdate <= curdate() AND termdate <> '0000-00-00';
```

The average length of employment for terminated employees is around 7.8 years.

### 8. Which department has the highest turnover rate?

```python
%%sql

WITH department_terminated AS (
SELECT
	department,
	COUNT(*) AS total_count,
    SUM(CASE WHEN termdate = '0000-000-00' THEN 1 ELSE 0 END) AS active_count,
    SUM(CASE WHEN termdate <> '0000-00-00' AND termdate <= curdate() THEN 1 ELSE 0 END) AS terminated_count
FROM
	hr
GROUP BY 
	department
)
SELECT
	department,
    total_count,
    active_count,
    terminated_count,
    terminated_count / total_count AS termination_rate
FROM
	department_terminated
ORDER BY
	termination_rate DESC;
```

The Auditing department has the highest turnover rate followed by Legal. The least turn over rate are in the Marketing departments.

### 10. How has the company's employee count changed over time based on hire and term dates?

```python
%%sql
WITH terminations_hires AS(
SELECT
	YEAR(hire_date) AS year,
    COUNT(*) AS hires,
    SUM(CASE WHEN termdate <> 0 AND termdate < CURDATE() THEN 1 ELSE 0 END) AS terminations
FROM 
	hr
GROUP BY
	year
ORDER BY
	year ASC
)
SELECT 
	year,
    hires,
    terminations,
    hires - terminations AS  net_change,
    ROUND((hires - terminations)/hires * 100 ,2) AS net_change_percent
FROM
	terminations_hires;
```

=> The net change in employees has increased over the years.

## 11. What is the tenure distribution for each department?

```python
%%sql


SELECT
	department,
    ROUND(AVG(datediff(termdate,hire_date)/365),0) AS avg_tenure
FROM 
	hr
WHERE
	termdate <> '0000-00-00' AND termdate <= curdate()
GROUP BY
	department
ORDER BY 
    avg_tenure DESC;
```

=> The average tenure for each department is about 8,9 years with Sales having the highest and Product Management, Legal having the lowest.
